## Image cropping

### Inport libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Read RGB Image

```python
img=cv2.imread('rgb_img.png')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```
