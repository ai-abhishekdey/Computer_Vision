## Theorytically Conversion of RGB to GRAY 

```

Converting a color image to grayscale involves reducing the three color channels (Red, Green, and Blue) to a single channel that represents the intensity of light. This conversion is typically done using a weighted sum of the R, G, and B values, reflecting the human eye's sensitivity to different wavelengths of light. The formula used in this conversion is:

Gray=0.299×R+0.587×G+0.114×B

Here’s a breakdown of why these specific weights are used:

Red (R) contributes 29.9% to the intensity because the human eye is less sensitive to red light compared to green.
Green (G) contributes 58.7% to the intensity as the human eye is most sensitive to green light.
Blue (B) contributes 11.4% to the intensity since the human eye is least sensitive to blue light.

```

````python
def rgb_to_gray(image,x,y):
    
    r,g,b = image[x,y]
    
    gray_value = 0.299 * r + 0.587 * g + 0.114 * b
    
    
    return int(gray_value)
````

````python
print("Red      :", rgb_to_gray(img,0,0))
print("Green    :", rgb_to_gray(img,0,40))
print("Blue     :", rgb_to_gray(img,0,60))
print("Black    :", rgb_to_gray(img,0,80))

print("Yellow   :", rgb_to_gray(img,80,0))
print("Cyan     :", rgb_to_gray(img,80,40))
print("Magenta  :", rgb_to_gray(img,80,60))
print("White    :", rgb_to_gray(img,80,80))
````
