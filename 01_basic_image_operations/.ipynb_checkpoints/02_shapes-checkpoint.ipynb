### Import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Display function

```python
def show_img(image, title):

    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
```

### Read image and convert to RGB

```python
bgr_image = cv2.imread("parrots.jpg")
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)


show_img(rgb_image, "rgb image")
```

### Create a Black Canvas or black background image

```python
black_canvas = np.zeros((400,400,3), dtype=np.uint8)

show_img(black_canvas, "black canvas")
```

```python
black_canvas.shape
```

### Create a white Canvas or white background image

```python
white_canvas = np.ones((480,640,3), dtype=np.uint8)

show_img(white_canvas, "white canvas")
```

#### NOTE: By default Height x Width Convention is considered


### Draw a rectangle on the black canvas

```python
## Syntax : cv2.rectangle(image, start_point, end_point, color, thickness)


start_point= (20,30)
end_point = (100,200)
color = (255,0,0)
thickness = 5


rectangle = cv2.rectangle(
    black_canvas,
    start_point,
    end_point,
    color,
    thickness
)


show_img(rectangle, "rectangle on black canvas")
```
