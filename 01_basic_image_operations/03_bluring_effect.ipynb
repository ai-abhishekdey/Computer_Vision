### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Display Function

```python
def show_img(image,title):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
```

### Read image and convert to RGB

```python
image = cv2.imread("parrots.jpg")
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

show_img(rgb_image, "rgb image")
```

### Gaussian Blur

```python
kernel=19 # Kernel size should be odd number

gauss_blur = cv2.GaussianBlur(rgb_image,(kernel,kernel),0)

show_img(gauss_blur,"gauss blur")
```

### Median Blur

```python
kernel=19

median_blur = cv2.medianBlur(rgb_image, kernel)

show_img(median_blur, "median blur")
```

### Bilateral filter

```python
bilateral_img = cv2.bilateralFilter(rgb_image, 40, 90,90)

show_img(bilateral_img, "bilateral filter")
```
