### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read image

```python
image=cv2.imread("parrots.jpg")
```

```python
image
```

### Image Dimensions

```python
image.shape
```

### Display image in window using opencv

```python
cv2.imshow("3 channel RGB image)",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Display image in Notebook using matplotlib

```python
def show_img(img,title,cmap=None):
    plt.imshow(img,cmap=cmap)
    plt.title(title)
    plt.axis('off')
```

### Note: By default open cv uses BGR convention

### Display default BGR image

```python
show_img(image,"BGR image")
```

### Display only Blue Channel (index 0)

```python
'''
cv2.imshow("Blue channel binary image",image[:,:,1])
cv2.waitKey(0)
cv2.destroyAllWindows()
'''

show_img(image[:,:,0], "Blue channel binary image")

```

### Display only Green Channel (index 1)

```python
'''
cv2.imshow("Green channel binary image",image[:,:,1])
cv2.waitKey(0)
cv2.destroyAllWindows()
'''

show_img(image[:,:,1], "Green channel binary image")
```

### Display only Red Channel (index 2)

```python
'''
cv2.imshow("Red channel binary image",image[:,:,2])
cv2.waitKey(0)
cv2.destroyAllWindows()
'''


show_img(image[:,:,2], "Red channel binary image")
```

### Convert BGR to RGB 

```python
rgb_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
show_img(rgb_image, "RGB image")
```

### Convert BGR to GRAY 

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
show_img(gray_image, "Grayscale image",cmap='gray')
```

### Resize image

```python
resized_image = cv2.resize(rgb_image,(64,32))

show_img(resized_image," Resized image")
```

```python
resized_image.shape
```

### Crop Image

```python
rgb_image.shape
```

```python
cropped_image = rgb_image[50:200,100:300]

show_img(cropped_image, "Cropped image")
```

```python
cropped_image.shape
```

### Rotate image

```python
rotated_image_90_clockwise=cv2.rotate(rgb_image,cv2.ROTATE_90_CLOCKWISE)
rotated_image_90_counter_clockwise=cv2.rotate(rgb_image,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotated_image_180=cv2.rotate(rgb_image,cv2.ROTATE_180)
```

```python
show_img(rotated_image_90_clockwise, "rotated_image_90_clockwise")
```

```python
show_img(rotated_image_90_counter_clockwise, "rotated_image_90_counter_clockwise")
```

```python
show_img(rotated_image_180, "rotated_image_180")
```

### Flip Image

```python
flipped_image=cv2.flip(rgb_image,1)

show_img(flipped_image, " flipped image")
```

### Gaussian Blur

```python
blur_image = cv2.GaussianBlur(rgb_image,(15,15),0)

show_img(blur_image, "blur_image")
```
