### Import libraries

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

### Display Function

````python
def show_img(image,title,cmap=None):
    plt.imshow(image,cmap=cmap)
    plt.title(title)
    plt.axis('off')
````

### Read Image and convert into gray scale

````python
image=cv2.imread('rectangles.png')
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
show_img(gray_image,'gray image', cmap='gray')
````

## Sobel edge detector


```
dst = cv2.Sobel(src, ddepth, dx, dy, ksize)

```


* src: The input image (typically grayscale).

* ddepth: The desired depth of the destination image. A common choice for 8-bit source images is cv2.CV_64F

* dx,dy=0,1 # for horizontal edges

* dx,dy=1,0 # for vertical egdes

* ksize: kernel size

### Sobel - horizontal edge detector

````python
sobel_img_h = cv2.Sobel(gray_image,cv2.CV_64F,0,1,21)
show_img(sobel_img_h,'sobel - horizontal edge',cmap='gray')
````

### Sobel - vertical edge detector

````python
sobel_img_v = cv2.Sobel(gray_image,cv2.CV_64F,1,0,21)
show_img(sobel_img_v,'sobel - vertical edge',cmap='gray')
````

````python


# Convert to 8-bit unsigned integers
abs_sobel_img_h = cv2.convertScaleAbs(sobel_img_h)
abs_sobel_img_v = cv2.convertScaleAbs(sobel_img_v)

# Combine both directions
sobel_img_hv = cv2.addWeighted(abs_sobel_img_h, 0.5, abs_sobel_img_v, 0.5, 0)

show_img(sobel_img_hv,'sobel - horizontal + vertical edge',cmap='gray')
````

## Canny edge detector (Most Popular)

````python
canny_edge_img = cv2.Canny(gray_image, threshold1=100, threshold2=200)

show_img(canny_edge_img,'Canny edge detection',cmap='gray')
````

## Laplacian Edge detection

````python
laplacian_edge_img = cv2.Laplacian(gray_image, cv2.CV_64F)
laplacian_edge_img_abs = cv2.convertScaleAbs(laplacian_edge_img)

show_img(laplacian_edge_img_abs,'Laplacian edge detection',cmap='gray')
````
